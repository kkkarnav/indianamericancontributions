```python
import pandas as pd
```

```python
population = pd.read_csv("./data/aapi_population.csv")
population
```

```python
yearly_sums = population.groupby("Year")["Indian Pop"].sum()
yearly_sums
```

```python
population["Indian Share"] = population.apply(
    lambda row: row["Indian Pop"] / yearly_sums[row["Year"]] * 100, axis=1
)
population
```

```python
population.sort_values(by="Indian Share", ascending=False).iloc[:10]
```
